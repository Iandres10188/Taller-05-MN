```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
xs = np.array([
    0.0003, 0.0822, 0.2770, 0.4212, 0.4403, 0.5588, 0.5943, 0.6134, 0.9070, 1.0367,
    1.1903, 1.2511, 1.2519, 1.2576, 1.6165, 1.6761, 2.0114, 2.0557, 2.1610, 2.6344
])
ys = np.array([
    1.1017, 1.5021, 0.3844, 1.3251, 1.7206, 1.9453, 0.3894, 0.3328, 1.2887, 3.1239,
    2.1778, 3.1078, 4.1856, 3.3640, 6.0330, 5.8088, 10.5890, 11.5865, 11.8221, 26.5077
])

# Ajuste grado 2
coef2 = np.polyfit(xs, ys, 2)
poly2 = np.poly1d(coef2)

# Ajuste grado 3
coef3 = np.polyfit(xs, ys, 3)
poly3 = np.poly1d(coef3)

# Valores para graficar
x_plot = np.linspace(min(xs), max(xs), 500)
y2_plot = poly2(x_plot)
y3_plot = poly3(x_plot)

# Gráfica
plt.figure(figsize=(10, 6))
plt.scatter(xs, ys, color='red', label='Datos originales')
plt.plot(x_plot, y2_plot, color='blue', linestyle='--', label='Ajuste grado 2')
plt.plot(x_plot, y3_plot, color='green', linestyle='-', label='Ajuste grado 3')
plt.title("Ajuste por mínimos cuadrados (grados 2 y 3)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Mostrar ecuaciones en consola
print("Polinomio grado 2:\n", poly2)
print("\nPolinomio grado 3:\n", poly3)

# Mostrar gráfica
plt.show()

```
