```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
xs = np.array([
    0.0003, 0.0822, 0.2770, 0.4212, 0.4403,
    0.5588, 0.5943, 0.6134, 0.9070, 1.0367,
    1.1903, 1.2511, 1.2519, 1.2576, 1.6165,
    1.6761, 2.0114, 2.0557, 2.1610, 2.6344,
])
ys = np.array([
    1.1017, 1.5021, 0.3844, 1.3251, 1.7206,
    1.9453, 0.3894, 0.3328, 1.2887, 3.1239,
    2.1778, 3.1078, 4.1856, 3.3640, 6.0330,
    5.8088, 10.5890, 11.5865, 11.8221, 26.5077,
])

# Ajuste cuadrático
coef = np.polyfit(xs, ys, 2)
print("Coeficientes del polinomio:", coef)

# Evaluación
x_fit = np.linspace(min(xs), max(xs), 300)
y_fit = np.polyval(coef, x_fit)

# Gráfica
plt.figure(figsize=(8, 6))
plt.plot(xs, ys, 'o', label='Datos')
plt.plot(x_fit, y_fit, '-', label='Ajuste cuadrático')
plt.title('Ajuste por mínimos cuadrados')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
